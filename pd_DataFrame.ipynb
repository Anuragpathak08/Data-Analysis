# DataFrame<hr> 
- A pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). It is the primary data structure in the pandas library

1)  Creating the `DataFrame` using list

```python
l = [[100,90,'20'],
     [80,50,'3'],
     [120,80,'30',]]

df  = pd.DataFrame(l)
df
```

2. Creating the `DataFrame` using dictionary

```python
# using dicts

student_dict = {
    'name':['nitish','ankit','rupesh','rishabh','amit','ankita'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students
```

+ if we want to make a column as an index then we will use `DataFrame.set_index('column_name', inplace= True)`

```python
students.set_index('name',inplace=True)
students
```
