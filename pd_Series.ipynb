# Series<hr>
A Pandas Series is a one-dimensional labeled array in Python's Pandas library capable of holding data of any type (integers, strings, floats, etc.). It is similar to a single column in a spreadsheet or a Python list, but with an explicitly defined index that provides additional functionality for data manipulation and analysis

To create a pandas series we will use the following code after importing pandas as pd 
```pd.Series(data)```

 here data can be a list or numpy array , etc

````python
country = ["India", 'pakistan', 'USA', 'Germany']

seriers = pd.Series(country)
seriers
````

````python
runs = [13,24,57, 46, 100, 43 , 12]
pd.Series(runs)
````

## Custom index in pandas<hr>
we can have custom index instead of  regular index (e.g. : 1,2,3,4)

````python
marks = [55, 68, 42, 93, 80]
subjects = ['Hindi', 'English', 'Marathi', 'maths', 'science' ]

pd.Series(marks, index= subjects)
````

## Indexing<hr>
Indexing in a pandas Series allows you to access and select specific elements using either position-based indexing (like standard Python lists) or label-based indexing (using custom labels) (**syntax** : ``` variable[index] ``` )

````python
age = pd.Series([10, 12, 44, 65, 21], index=['rahul', 'ankit', 'pankaj', 'rusih', 'aman'], name = 'age')
age['rahul']
````

## Slicing<hr>
Slicing a series in Python (typically a Pandas Series object) is a powerful way to extract specific subsets of data. You can slice by position using integer indices or by label using the series' index label (**syntax** : ```variable[start index : end index : updation] ``` )

````python
age[-1:1:-1]
````

## Filtering<hr> 
You can filter a pandas Series using boolean indexing (filtering by value based on a condition)
This is the most common method, allowing you to select elements based on their values. Applying a condition to the Series generates a boolean Series (True/False mask), which is then used to select the corresponding data.

**syntax** : ``` variable[condition] ```

````python
age[age>18]
````

- ### Indexing in Series <hr>
Indexing in a pandas Series is the mechanism used to access, identify, and align individual data elements using associated labels or integer positions. The index acts as an "address" for the data, enabling efficient retrieval and manipulation.

````python
# 1. Integer Index (Label Priority)
s_int = pd.Series([10, 20, 30], index=[0, 1, 2])
s_int[1]    # 20 (Label)
# s_int[-1]   # KeyError (No label -1)
````

### The Core Problem
When using `s[]`, Pandas treats the value as a **Label ONLY**.

👉 No more ambiguity:
**There is NO position fallback anymore.**<hr>

### The Rules of `s[]` (Updated)

| Index Type | `s[1]` Behavior | `s[-1]` Behavior |
| :--- | :--- | :--- |
| **Integer Index**<br>`[0, 1, 2]` | ✅ **Label** lookup<br>(value at label `1`) | ❌ **KeyError**<br>(no label `-1`) |
| **Non-Integer Index**<br>`['a', 'b', 'c']` | ❌ **KeyError**<br>(no label `1`) | ❌ **KeyError**<br>(no label `-1`) |
<hr>

### Why This Happens

1. **Strict Label Behavior:**  
   `s[x]` always treats `x` as a **label**

2. **No Fallback:**  
   Pandas **does NOT fallback to position anymore**

3. **Consistency Goal:**  
   Matches DataFrame behavior → avoids silent bugs

---


````python
# 2. String Index (Position Fallback)
s_str = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
# s_str[0] # KeyError (No label 0)
s_str['a']
````
